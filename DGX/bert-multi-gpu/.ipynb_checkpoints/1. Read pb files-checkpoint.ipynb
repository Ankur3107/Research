```python
import tensorflow as tf
from pathlib import Path
from tensorflow.contrib import predictor
import numpy as np
```

```python
export_dir = 'bert/export'
subdirs = [x for x in Path(export_dir).iterdir()
               if x.is_dir() and 'temp' not in str(x)]
```

```python
latest = str(sorted(subdirs)[-1])
```

```python
latest
```

```python
predict_fn = predictor.from_saved_model(latest)
```

```python
predict_fn
```

```python
unique_ids = [1020203]
segment_ids =  np.array([0]*384).reshape(1,384)
input_ids = np.array([0]*384).reshape(1,384)
input_mask = np.array([0]*384).reshape(1,384)
```

```python
predictions = predict_fn({'unique_ids': unique_ids, 'segment_ids': segment_ids, 'input_ids':input_ids, 'input_mask':input_mask})
```
