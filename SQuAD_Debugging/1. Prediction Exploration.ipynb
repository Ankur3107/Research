```python
import pandas as pd
import json
```

```python
data_file = '/Users/ankur.kumar/Downloads/albert_backup/SQuAD/dev-v2.0.json'
```

```python
with open(data_file) as f:
    dataset_json = json.load(f)
    dataset = dataset_json['data']
```

```python
dataset[0]
```

```python
def get_answer(ls):
    result = []
    for l in ls:
        result.append(l['text']) 
    return result

dataset_dict = {}

for ds in dataset:
    for ds_para in ds['paragraphs']:
        for ds_para_qas in ds_para['qas']:
            dataset_dict[ds_para_qas['id']] = get_answer(ds_para_qas['answers'])
```

```python
dataset_dict
```

```python
pred_file_single = '/Users/ankur.kumar/Desktop/predictions_single.json'
pred_file_ensemble = '/Users/ankur.kumar/Desktop/predictions_ensemble.json'
```

```python
with open(pred_file_single) as f:
    preds_single = json.load(f)
    
with open(pred_file_ensemble) as f:
    preds_ensemble = json.load(f)
```

```python
preds_single
```

## Predictions EDA

### 1. How many No Answer?

```python
count = 0
for key,value in dataset_dict.items():
    if len(value)==0:
        count+=1
        
print('Total No Answer (Actual):', count)
```

```python
count_single = 0
for key,value in preds_single.items():
    if value=='':
        count_single+=1
        
print('Total No Answer (Single Model Prediction) :', count_single)
```

```python
count_ensemble = 0
for key,value in preds_ensemble.items():
    if value=='':
        count_ensemble+=1
        
print('Total No Answer (Single Model Prediction) :', count_ensemble)
```

```python
print('Diff in Single Model Prediction and Actual :',count-count_single)
print('Diff in Ensemble Model Prediction and Actual :',count-count_ensemble)
```

### 2. How many Answer mismatch?

```python
import evaluation
```

```python
exact_count_single = 0
for id, gold_answers in dataset_dict.items():
    a_pred = preds_single[id]
    
    if len(gold_answers) == 0 and a_pred=='':
        exact_count_single+=1
        continue
    elif len(gold_answers) == 0 and a_pred!='':
        continue
    exact_count_single+=max(evaluation.compute_exact(a, a_pred) for a in gold_answers)
    
print('Total mismatch using Single Model Prediction :', len(dataset_dict)-exact_count_single,' & Match Percentage :',exact_count_single/len(dataset_dict))
```

```python
exact_count_ensemble = 0
for id, gold_answers in dataset_dict.items():
    a_pred = preds_ensemble[id]
    
    if len(gold_answers) == 0 and a_pred=='':
        exact_count_ensemble+=1
        continue
    elif len(gold_answers) == 0 and a_pred!='':
        continue
    exact_count_ensemble+=max(evaluation.compute_exact(a, a_pred) for a in gold_answers)
    
print('Total mismatch using Single Model Prediction :', len(dataset_dict)-exact_count_ensemble,' & Match Percentage :',exact_count_ensemble/len(dataset_dict))
```

### No Answer Match 

```python
exact_no_answer_count_single = 0
wrong_no_answer_count_single = 0
for id, gold_answers in dataset_dict.items():
    a_pred = preds_single[id]
    
    if len(gold_answers) == 0 and a_pred=='':
        exact_no_answer_count_single+=1
        continue
    elif len(gold_answers) == 0 and a_pred!='':
        wrong_no_answer_count_single+=1
        continue

print('Total Match & Mismatch No Answer using Single Model Prediction :', exact_no_answer_count_single,wrong_no_answer_count_single)
```

```python
exact_no_answer_count_ensemble = 0
wrong_no_answer_count_ensemble = 0
for id, gold_answers in dataset_dict.items():
    a_pred = preds_ensemble[id]
    
    if len(gold_answers) == 0 and a_pred=='':
        exact_no_answer_count_ensemble+=1
        continue
    elif len(gold_answers) == 0 and a_pred!='':
        wrong_no_answer_count_ensemble+=1
        continue

print('Total Match & Mismatch No Answer using Ensemble Model Prediction :', exact_no_answer_count_ensemble,wrong_no_answer_count_ensemble)
```

## Mis-Match Depth Analysis

```python
mismatch_single_dict = {}
for id, gold_answers in dataset_dict.items():
    a_pred = preds_single[id]
    
    if len(gold_answers) == 0 and a_pred=='':
        exact_count_single+=1
        continue
    elif len(gold_answers) == 0 and a_pred!='':
        mismatch_single_dict[id] = a_pred
        continue
    elif max(evaluation.compute_exact(a, a_pred) for a in gold_answers) != 1:
        mismatch_single_dict[id] = a_pred
```
